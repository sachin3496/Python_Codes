```python
from numpy import * 
from pylab import * 
```

```python
a = [ 1,2,3,4,5,6]
b = [ 7,8,9,10,11,12]
a = array(a)
b = array(b)
print(type(a))
print(type(b))
print(a)
print(b)
```

```python
print(a+b)
print(a*b) #element vise multiplicaton
print(a@b)#dot multiplication
print(b/a)
print(3*a)
```

```python
s = tan(a)
print(s)
```

```python
x = [ (var*pi)/180 for var in range(361) ]
a= sin(array(x))
b = cos(array(x))
plot(a)
plot(b)
show()
```

```python
x = [ (var*pi)/180 for var in range(361) ]
x.pop(90)
x.pop(269)
a = sin(x)
b  = cos(x)

plot(a/b)
show()
```

```python
x = [ (var*pi)/180 for var in range(75) ]
x  = x + [ (var*pi)/180 for var in range(110,250) ]
x = x + [ (var*pi)/180 for var in range(280,361) ]
a = sin(x)
b  = cos(x)

plot(a/b)
show()
```
