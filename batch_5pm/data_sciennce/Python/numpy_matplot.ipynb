```python
angles = np.arange(0,361,15)
print(angles)
```

```python
new_angles = (angles*np.pi)/180
sin  = np.sin(new_angles)
cos  = np.cos(new_angles)
```

```python
# relation relation 
plt.plot(angles,sin,'r:',linewidth=2,label='sin')
plt.plot(angles,cos,'b:',linewidth=2,label='cos')
plt.title('sin&cos')
plt.legend()
plt.xlabel('angles')
plt.ylabel('values')
plt.show()
```
