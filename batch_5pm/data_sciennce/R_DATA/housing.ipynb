```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.random((5,5))
y = np.random.random((6,6))
print(x)
print(y)
```

```python
plt.plot(x,label='X',color='red',ls='--')
plt.plot(y,label='y',color='green',ls='solid')
plt.legend()
plt.show()
```

```python
x = [ var for var in range(360)]
y = x.copy()
y.remove(90)
y.remove(270)
degree = np.pi/180
x = np.array(x)
y = np.array(y)
x = x*degree
y  = y*degree
```

```python
x_sin = np.sin(x)
y_cos = np.cos(x)
z_tan = np.tan(y)
```

```python
plt.plot(x_sin,color='red',label='sin')
plt.plot(y_cos,color='green',label='cos')
plt.legend()
plt.savefig('sin_cos.png')
plt.show()
plt.plot(z_tan,color='yellow',label='tan',)
plt.show()
```
